```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var1 = [1,2,3,4,5,6,7]
var2 = [3,5,2,1,5,6,7]

x_1 = pd.DataFrame({"var1": var1, "var2": var2})
```

```python
plt.plot(var1, var2)
plt.show()
```

```python
sns.lineplot(x = "var1", y = "var2", data=x_1)
plt.show()
```
