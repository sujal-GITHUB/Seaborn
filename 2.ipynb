```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y_1 = sns.load_dataset('diamonds')
y_1
```

```python
sns.lineplot(x = "carat", y = "clarity", data = y_1 )
plt.show()
```

```python
sns.lineplot(x = "carat", y = "clarity", data = y_1, hue="color")
plt.show()
```

```python
sns.lineplot(x = "carat", y = "clarity", data = y_1, hue="color", style="cut", palette="Accent")
plt.show()
```
